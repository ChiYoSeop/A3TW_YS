```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

df = pd.read_csv('../results/csvs/AnalysisResult_A3.csv',encoding = 'CP949')

dflst = df.columns.tolist()

warnings.filterwarnings('ignore')

fig, ax = plt.subplots(ncols=2, nrows=4, figsize = (20,40))
plt.suptitle('Wafer-To-Wafer Analysis',fontsize = 50, y= 0.9)
plt.rcParams['ytick.minor.visible'] = True    # y축 보조 눈금
plt.rcParams['font.size'] = 20

def grph(a,b,c):
    sw = sns.swarmplot(data = df, x = "Wafer", y = dflst[a], ax = ax[b,c])
    box = sns.boxplot(data = df, x = "Wafer", y = dflst[a], showcaps = False, # 박스 상단 가로라인 보이지 않기
                whiskerprops = {'linewidth':0}, # 박스 상단 세로 라인 보이지 않기 
                showfliers = False,           # 박스 범위 벗어난 아웃라이어 표시하지 않기
                boxprops = {'facecolor':'None'}, ax = ax[b,c])
    return sw,box

grph(10,0,0)
for i in range(12,15):
    grph(i,i-11,0)
for i in range(15,19):
    grph(i,i-15,1)
    
plt.savefig('../results/wafer-to-wafer/Wafer-to-Wafer Analysis.png', dpi = 300, facecolor = '#eeeeee')
```
